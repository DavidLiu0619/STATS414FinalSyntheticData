```python
import pandas as pd
from sdv.metadata.metadata import Metadata


# Load your CSV data
real_data = pd.read_csv('train_ads_minimal_label1.csv')

# Define metadata for your single-table dataset
metadata_dict = {
    'columns': {
        "label": {"sdtype": "categorical"},
        "age": {"sdtype": "categorical"},
        "gender": {"sdtype": "categorical"},
        "residence": {"sdtype": "categorical"},
        "city_rank": {"sdtype": "categorical"},
        "series_dev": {"sdtype": "categorical"},
        "series_group": {"sdtype": "categorical"},
        "emui_dev": {"sdtype": "categorical"},
        "net_type": {"sdtype": "categorical"},
        "creat_type_cd": {"sdtype": "categorical"},
        "inter_type_cd": {"sdtype": "categorical"},
        "slot_id": {"sdtype": "categorical"},
        "hispace_app_tags": {"sdtype": "categorical"},
        "app_second_class": {"sdtype": "categorical"},
        "app_score": {"sdtype": "categorical"},
        "u_refreshTimes": {"sdtype": "categorical"},
        "u_feedLifeCycle": {"sdtype": "categorical"}
    },
    'primary_key': None,
    'alternate_keys': [],
    'sequence_key': None,
    'sequence_index': None,
    'METADATA_SPEC_VERSION': 'SINGLE_TABLE_V1'
}

# Load metadata using the new Metadata class
metadata = Metadata.load_from_dict(metadata_dict)

# Extract the table metadata (for single table, it's stored under 'table')
table_metadata = metadata.tables['table']

# Initialize the TVAE synthesizer with the extracted metadata
#synthesizer = TVAESynthesizer(table_metadata)
synthesizer = TVAESynthesizer(
    table_metadata, # required
    enforce_min_max_values=True,
    enforce_rounding=True,
    epochs=500
)

# Fit the synthesizer on the real data
synthesizer.fit(real_data)

# Generate synthetic data (e.g., 10 rows)
synthetic_data = synthesizer.sample(num_rows=119136)
```

```python
synthetic_data
```

```python
synthetic_data.to_csv('synthetic_data_TVAE_Epoch10.csv', index=False)
```
