# GReaT Synthetic Data

# For Ads Datasets

```python
import pandas as pd
data = pd.read_csv("train_ads_minimal_label1.csv")  #119136 x 17
print(data.shape)
```

# Batch Size for the 1000

```python
from be_great import GReaT
import pandas as pd
# Initialize the GReaT model
# Using distilgpt2 which has a context window of ~1024 tokens.
model = GReaT(llm='distilgpt2', batch_size=64, epochs=2, save_steps=400000)

# Train the model on your dataset
model.fit(data)
```

```python
model.save("GReaT_model")
```

```python
# Define total number of synthetic samples and the batch size for generation
total_samples = 119000 #119136
batch_size_gen = 1000  # Adjust this based on your available memory and resources
all_synthetic = {}
# Generate synthetic data in batches
num_batches = total_samples // batch_size_gen  # Should be 2 in this case

# Generate synthetic data in batches using a max_length that fits the model's context window
for i in range(0, total_samples, batch_size_gen):
    start = i * batch_size_gen
    end = start + batch_size_gen
    print(f"Generating samples {i} to {i + batch_size_gen}...")
    batch = model.sample(max_length=1024, n_samples=batch_size_gen)
    # Store the batch DataFrame in the dictionary with a unique key
    batch_df = pd.DataFrame(batch)
    all_synthetic[f"batch_{i + 1}"] = batch_df
```

```python
combined_synthetic_df = pd.concat(all_synthetic.values(), ignore_index=True)
print("Combined DataFrame shape:", combined_synthetic_df.shape)
```

```python
combined_synthetic_df.to_csv('synthetic_data_GReaT119000.csv', index=False)
```
